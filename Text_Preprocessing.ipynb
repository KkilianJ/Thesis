```python
import pandas as pd
import re
import emoji
from joblib import Parallel, delayed
from tqdm import tqdm

def clean_tweet(text):
    text = emoji.demojize(text)
    text = re.sub(r"http\S+", "", text, flags=re.IGNORECASE)
    text = re.sub(r"www\S+", "", text, flags=re.IGNORECASE)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#\w+", "", text)
    text = re.sub(r"[^\w\s:]", "", text)
    text = text.lower()
    text = re.sub(r"\s+", " ", text).strip()
    return text

def process_file(input_path, output_path):
    df = pd.read_csv(input_path, low_memory=False)
    texts = df["text"].astype(str).tolist()
    cleaned = Parallel(n_jobs=12)(
        delayed(clean_tweet)(t) for t in tqdm(texts, desc="Cleaning")
    )
    df["bigram_text"] = cleaned
    df.to_csv(output_path, index=False)

b = "/content/drive/MyDrive/Thesis/"
process_file(b + "long_text_sample.csv", b + "long_text_with_bigram_text.csv")
process_file(b + "short_text_sample.csv", b + "short_text_with_bigram_text.csv")
```

```python
import pandas as pd
import re
import emoji
from joblib import Parallel, delayed
from tqdm import tqdm
import re

def clean_tweetpos(text):
    text = re.sub(r"[\U00010000-\U0010ffff]", "", text) #delete emoji
    text = re.sub(r"http\S+", "", text, flags=re.IGNORECASE)
    text = re.sub(r"www\S+", "", text, flags=re.IGNORECASE)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#\w+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = text.lower()
    text = re.sub(r"\s+", " ", text).strip()
    return text

def process_file(input_path, output_path):
    df = pd.read_csv(input_path, low_memory=False)
    texts = df["text"].astype(str).tolist()
    cleaned = Parallel(n_jobs=12)(
        delayed(clean_tweetpos)(t) for t in tqdm(texts, desc="Cleaning")
    )
    df["bigram_text"] = cleaned
    df.to_csv(output_path, index=False)

b = "/content/drive/MyDrive/Thesis/"
process_file(b + "long_text_sample.csv", b + "long_text_with_pos_text.csv")
process_file(b + "short_text_sample.csv", b + "short_text_with_pos_text.csv")
```
